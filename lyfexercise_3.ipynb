```python
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords, gutenberg
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
import matplotlib.pyplot as plt
```

```python
nltk.download('gutenberg')
```

```python
moby_dick_text = gutenberg.raw("melville-moby_dick.txt")
print(moby_dick_text[:100])
```

```python
# Step 2: Tokenization
tokens = word_tokenize(moby_dick_text.lower())
```

```python
# Step 3: Stop-words filtering
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token not in stop_words]
```

```python
# Step 4: Parts-of-Speech (POS) tagging and POS frequency
pos_tags = pos_tag(filtered_tokens)
pos_counts = FreqDist(tag for word, tag in pos_tags)
top_pos = pos_counts.most_common(5)
```

```python
print("Top 5 parts of speech and their counts:")
for pos, count in top_pos:
    print(f"{pos}: {count}")
```

```python

# Step 5: Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word, tag in pos_tags[:20]]
print("\nTop 20 lemmatized tokens:")
for token in lemmatized_tokens:
    print(token)
```

```python

# Step 6: Plotting frequency distribution
pos_labels, pos_counts = zip(*pos_counts.items())

plt.bar(pos_labels, pos_counts)
plt.xlabel('Part of Speech')
plt.ylabel('Frequency')
plt.title('POS Frequency Distribution')

plt.show()
```
